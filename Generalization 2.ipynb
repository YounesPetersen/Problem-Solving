```python
def sum_of_series(n,m):
    
    sum = 1/m
    for i in range(1, n//2):
        sum += 1/(m*i)
    return sum

sum_of_series(10,3)

```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_series_sum(k, l):
    plt.figure(figsize=(14, 8))

    # Iterate over m values from 1 to l
    for m in range(1, l+1):
        n_values = np.arange(1, k+1)
        sum_values = np.array([sum_of_series(n, m) for n in n_values])
        first_n_greater_than_1 = next((i for i, sum_val in enumerate(sum_values) if sum_val > 1), None)

        # Plotting the sum series for each m
        plt.plot(n_values, sum_values, marker='o', markersize=1, label=f"m={m}")

        # Annotating the first point where sum is greater than 1 for each m
        if first_n_greater_than_1 is not None:
            print(f" m = {m}, n={n_values[first_n_greater_than_1]}, Sum={sum_values[first_n_greater_than_1]:.4f}")

    plt.axhline(y=1, color='r', linestyle='-', label="1")
    plt.title("Sum of Series for m from 1 to 10")
    plt.xlabel("n")
    plt.ylabel("Sum")
    plt.legend()
    plt.grid(True)
    plt.show()

plot_series_sum(10000, 10)
```
