```python
def sum_of_series(n,m):
    
    sum = 0
    for i in range(1, n+1):
        sum += 1/(2*m*i)
    return sum

sum_of_series(11,2)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_sum_of_series(k,m):
    # Generating the range of n values from 1 to k
    n_values = np.arange(1, k+1)
    # Calculating the sum for each n value
    sum_values = np.array([sum_of_series(n,m) for n in n_values])

    # Finding the first n where the sum is greater than 0.75
    first_n_greater_than_075 = next((i for i, sum_val in enumerate(sum_values) if sum_val > 0.75), None)

    # Plotting the sum values
    plt.figure(figsize=(10, 6))
    plt.plot(n_values, sum_values, label="Sum of Series", marker='o')

    # Adding a horizontal line at y=0.75
    plt.axhline(y=0.75, color='r', linestyle='-', label="y = 0.75")

    if first_n_greater_than_075 is not None:
        plt.plot(n_values[first_n_greater_than_075], sum_values[first_n_greater_than_075], 'ro')
        plt.text(n_values[first_n_greater_than_075], sum_values[first_n_greater_than_075], 
                 f" n={n_values[first_n_greater_than_075]}, Sum={sum_values[first_n_greater_than_075]:.4f}")

    plt.title("Sum of 1/(4i) from 1 to n")
    plt.xlabel("n")
    plt.ylabel("Sum")
    plt.legend()
    plt.grid(True)
    plt.show()

plot_sum_of_series(2000,4)
```

```python
def plot_series_sum(k, l):
    plt.figure(figsize=(14, 8))

    # Iterate over m values from 1 to l
    for m in range(1, l+1):
        n_values = np.arange(1, k+1)
        sum_values = np.array([sum_of_series(n, m) for n in n_values])
        first_n_greater_than_075 = next((i for i, sum_val in enumerate(sum_values) if sum_val > 0.75), None)

        # Plotting the sum series for each m
        plt.plot(n_values, sum_values, marker='o', markersize=1, label=f"m={m}")

        # Annotating the first point where sum is greater than 0.75 for each m
        if first_n_greater_than_075 is not None:
            print(f" m = {m}, n={n_values[first_n_greater_than_075]}, Sum={sum_values[first_n_greater_than_075]:.4f}")

    # Common elements for all series
    plt.axhline(y=0.75, color='r', linestyle='-', label="y = 0.75")
    plt.title("Sum of Series for m from 1 to 10")
    plt.xlabel("n")
    plt.ylabel("Sum")
    plt.legend()
    plt.grid(True)
    plt.show()

plot_series_sum(10000, 10)
```
